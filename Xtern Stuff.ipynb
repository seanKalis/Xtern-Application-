```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("D:/Xtern.csv",converters={"power_level":int})
# The code above pulls in the CSV, and converts the values in column "power_levels" to integers for later calculation

op = open("D:/Xtern.csv")
next(op)
# The code above opens the file, and then jumps down a line to avoid counting the header

scooter_number = len(op.readlines())
full_charge = (scooter_number * 5)
current_charge = df["power_level"].sum()
need_to_charge = ((full_charge - current_charge)/5)
average_required_charge = (current_charge / scooter_number)
# The code above is declaring variables and their arguments

print (scooter_number, "- The number of scooters across the city")
print (full_charge, "- The full power capacity of all the scooters")
print (current_charge, "- Current power levels of all scooters")
print (need_to_charge, "- Number of hours to charge all scooters")
print (average_required_charge, "- Average charge required for all scooters")
# The code above prints the variables declared previously, as well as gives a descriptor as to what the variables mean
```
